Null hypothesis = there is no difference between median home value of objects with less than average rooms and those with more than average rooms 

Alternative hypothesis = there is a difference in price as price increases

Signifigance level = 0.5

```python
import pandas as pd 
from scipy.stats import ttest_rel
import requests

df = pd.read_csv("/content/sample_data/california_housing_test.csv")
df.head() # lets test if total rooms correlates to higher median house value 

targetDf = df[['total_rooms', 'median_house_value']] #removes other columns so we can focus on data we want to observe 

room_low = targetDf[targetDf['total_rooms'] < targetDf['total_rooms'].quantile(.25)] #rows that fit the constriction 

room_high = targetDf[targetDf['total_rooms'] > targetDf['total_rooms'].quantile(.75)] 

room_low.plot()

room_high.plot()

 


```

so here we can see that visual there is a bit higher median home value for the top 25% of total rooms compared to bottom 25%, lets investigate this with a paired t - test

```python
room_high = room_high[0:len(room_low.values)] #equaize the columns

ttest_rel(room_low['median_house_value'], room_high['median_house_value'])

```

Here we can see that we have far exceeded the signifigance level of 0.5 requeired to reject the null hypothesis
